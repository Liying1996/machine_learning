### 2. iris

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris

data = load_iris()

# type(data)
# 
print('')
print(data.target) 

# 
iris = pd.DataFrame(data=data.data, columns=data.feature_names)
print('')
print(iris.head(5)) 

# 
print('')
iris.info()
print('')
iris.describe()
```

### 3. iris
seaborn

```python
# iris_all
iris_all = iris.copy()
target = [data.target_names[i] for i in data.target]
iris_all['type'] = target
print(iris_all.head(5))

# pairplot
# help(sns.pairplot)
sns.pairplot(data=iris_all, hue='type') # density
sns.pairplot(data = iris_all, hue = 'type', diag_kind = 'hist')
```

```python
# featureboxplot
for name in iris_all.columns[:-1]:
    sns.boxplot(x='type', y=name, saturation=0.5,palette='pastel', data=iris_all)
    plt.title(name)
    plt.show()
```
